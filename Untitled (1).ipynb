```python
import os
import warnings
from transformers import logging

# Suppress Python warnings
warnings.filterwarnings("ignore")

# Suppress Transformers warnings/logs
logging.set_verbosity_error()

# Optional: silence HF + tokenizer parallel warnings
os.environ["TOKENIZERS_PARALLELISM"] = "false"
```

```python
import gc, random, torch
from datasets import load_dataset, Dataset
from sklearn.model_selection import train_test_split

from transformers import (AutoTokenizer, AutoModelForSeq2SeqLM,
                          M2M100ForConditionalGeneration, M2M100Tokenizer,
                          TrainingArguments, Trainer)

from peft import LoraConfig, get_peft_model

device = "cuda" if torch.cuda.is_available() else "cpu"

def clear_gpu():
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
```

```python
# Load
opus = load_dataset("opus100", "en-pa")
samanantar = load_dataset("ai4bharat/samanantar", "pa")

def extract_opus(ds):
    out = []
    for i in ds:
        en = i["translation"]["en"].strip()
        pa = i["translation"]["pa"].strip()
        if en and pa:
            out.append({"en": en, "pa": pa})
    return out

def extract_samanantar(ds):
    out = []
    for i in ds:
        en = i["src"].strip()
        pa = i["tgt"].strip()
        if en and pa:
            out.append({"en": en, "pa": pa})
    return out

opus_data = extract_opus(opus["train"])[:100000]
samanantar_data = extract_samanantar(samanantar["train"])[:100000]

data = opus_data + samanantar_data

# Light cleaning (helps BLEU/COMET + reduces hallucination)
clean = []
for x in data:
    if len(x["en"].split()) >= 3 and len(x["pa"].split()) >= 2:
        clean.append(x)

random.shuffle(clean)

train_data, val_data = train_test_split(clean, test_size=0.1, random_state=42)

print("Train:", len(train_data), "Val:", len(val_data))
```

```python
def preprocess(example, tokenizer):
    tokenizer.src_lang = "en"
    tokenizer.tgt_lang = "pa"   # ✅ FIX

    model_inputs = tokenizer(
        example["en"],
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    labels = tokenizer(
        example["pa"],
        max_length=128,
        truncation=True,
        padding="max_length"
    )["input_ids"]

    labels = [(t if t != tokenizer.pad_token_id else -100) for t in labels]

    model_inputs["labels"] = labels
    return model_inputs
```

```python
m2m_name = "facebook/m2m100_418M"

m2m_tokenizer = M2M100Tokenizer.from_pretrained(m2m_name)
m2m_model = M2M100ForConditionalGeneration.from_pretrained(m2m_name).to(device)

lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="SEQ_2_SEQ_LM"
)
m2m_model = get_peft_model(m2m_model, lora_config)

# Memory savers
m2m_model.config.use_cache = False
m2m_model.enable_input_require_grads()

m2m_model.print_trainable_parameters()
```

```python
train_ds = Dataset.from_list(train_data)
val_ds   = Dataset.from_list(val_data)

m2m_train = train_ds.map(lambda x: preprocess(x, m2m_tokenizer), remove_columns=["en","pa"])
m2m_val   = val_ds.map(lambda x: preprocess(x, m2m_tokenizer), remove_columns=["en","pa"])
```

```python
m2m_args = TrainingArguments(
    output_dir="./m2m_lora",
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    num_train_epochs=3,
    learning_rate=1e-4,

    evaluation_strategy="steps",
    eval_steps=1000,

    save_steps=2000,
    save_total_limit=2,

    fp16=False,
    logging_steps=200,
    report_to="none"
)
```

```python
m2m_trainer = Trainer(
    model=m2m_model,
    args=m2m_args,
    train_dataset=m2m_train,
    eval_dataset=m2m_val,
    tokenizer=m2m_tokenizer
)

m2m_trainer.train()

m2m_model.save_pretrained("./m2m_lora")
m2m_tokenizer.save_pretrained("./m2m_lora")

del m2m_model
clear_gpu()
```

```python
def preprocess_nllb(example):
    tokenizer.src_lang = "eng_Latn"
    tokenizer.tgt_lang = "pan_Guru"

    inputs = tokenizer(
        example["en"],
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    targets = tokenizer(
        example["pa"],
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    labels = targets["input_ids"]
    labels = [(t if t != tokenizer.pad_token_id else -100) for t in labels]

    inputs["labels"] = labels
    return inputs
```

```python
import torch
import pandas as pd
import matplotlib.pyplot as plt

from transformers import (
    M2M100ForConditionalGeneration,
    M2M100Tokenizer,
    AutoTokenizer,
    AutoModelForSeq2SeqLM
)

from peft import PeftModel

from sentence_transformers import SentenceTransformer, util
from bert_score import score
from nltk.translate.bleu_score import sentence_bleu
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
embed_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
nllb_back_tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
nllb_back_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M").to(device)
```

```python
nllb_tokenizer.src_lang = "pan_Guru"
```

```python
def back_translate_nllb(text_pa):
    nllb_back_tokenizer.src_lang = "pan_Guru"

    inputs = nllb_back_tokenizer(
        text_pa,
        return_tensors="pt",
        truncation=True,
        max_length=128
    ).to(device)

    with torch.no_grad():
        tokens = nllb_back_model.generate(
            **inputs,
            forced_bos_token_id=nllb_back_tokenizer.convert_tokens_to_ids("eng_Latn"),
            num_beams=5,
            max_length=128
        )

    return nllb_back_tokenizer.decode(tokens[0], skip_special_tokens=True)
```

```python
print(back_translate("ਮੀਟਿੰਗ ਕੱਲ੍ਹ ਨੂੰ ਮੁਲਤਵੀ ਕਰ ਦਿੱਤੀ ਗਈ ਹੈ।"))
```

```python
def direct_similarity(src_en, pa_text):
    emb1 = embed_model.encode(src_en, convert_to_tensor=True)
    emb2 = embed_model.encode(pa_text, convert_to_tensor=True)
    return util.cos_sim(emb1, emb2).item()
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
smooth = SmoothingFunction().method1

def evaluate_batch(src_list, pa_list):
    # Back translate (still per sentence, model can't batch easily)
    back_list = [back_translate_nllb(pa) for pa in pa_list]

    # 🔥 BATCH embeddings (FAST)
    emb_src = embed_model.encode(src_list, convert_to_tensor=True, batch_size=16)
    emb_pa  = embed_model.encode(pa_list, convert_to_tensor=True, batch_size=16)

    cosines = util.cos_sim(emb_src, emb_pa).diagonal().cpu().tolist()

    # BERTScore (already batched internally)
    P, R, F1 = score(back_list, src_list, lang="en", verbose=False)
    berts = F1.cpu().tolist()

    bleus = []
    for src, back in zip(src_list, back_list):
        bleu = sentence_bleu(
            [src.split()],
            back.split(),
            weights=(0.5, 0.5, 0, 0),
            smoothing_function=smooth
        )
        bleus.append(bleu)

    results = []
    for i in range(len(src_list)):
        final_score = 0.8 * berts[i] + 0.2 * cosines[i]

        results.append({
            "bleu": round(bleus[i], 4),
            "bert": round(berts[i], 4),
            "cosine": round(cosines[i], 4),
            "final_score": round(final_score, 4),
            "back_translation": back_list[i]
        })

    return results
```

```python
def adaptive_batch(texts):
    # Step 1: M2M translations (batch)
    m2m_outputs = [translate_m2m(t) for t in texts]

    # Step 2: Evaluate M2M (batch)
    scores_list = evaluate_batch(texts, m2m_outputs)

    final_results = []

    for i, text in enumerate(texts):
        scores = scores_list[i]

        # Decision
        if scores["final_score"] < 0.75 or scores["cosine"] < 0.30:
            nllb_out = translate_nllb(text)

            # Re-evaluate single (rare fallback → OK)
            new_scores = evaluate_batch([text], [nllb_out])[0]

            final_results.append({
                "input": text,
                "output": nllb_out,
                "model": "NLLB",
                **new_scores
            })
        else:
            final_results.append({
                "input": text,
                "output": m2m_outputs[i],
                "model": "M2M",
                **scores
            })

    return final_results
```

```python
from tqdm import tqdm

batch_size = 16
all_results = []

texts = [x["en"] for x in val_data[:200]]

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i+batch_size]
    res = adaptive_batch(batch)
    all_results.extend(res)
```
